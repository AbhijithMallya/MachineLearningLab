```python
def pname():
    print("4SF20CI002 - ABHIJITH MALLYA")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('election_data.csv')
pname()
print("Reading the csv file !")
```

```python
# Bar chart
pname()
gender_counts = data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
# Pie chart
pname()
age_ranges = pd.cut(data['age'], bins=[20, 30, 40, 50])
age_counts = age_ranges.value_counts()
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%')
plt.title('Age Distribution')
plt.axis('equal')
plt.show()
```

```python
# Scatter plot
pname()
plt.scatter(data['age'], data['votes'])
plt.xlabel('Age')
plt.ylabel('Votes')
plt.title('Age vs Votes')
plt.show()
```

```python
# Line plot
pname()
votes_by_id = data.groupby('id')['votes'].sum()
plt.plot(votes_by_id.index, votes_by_id.values)
plt.xlabel('ID')
plt.ylabel('Votes')
plt.title('Votes by ID')
plt.show()
```

```python
# Box plot
pname()
plt.boxplot(data['votes'])
plt.ylabel('Votes')
plt.title('Distribution of Votes')
plt.show()
```

```python
# Stacked bar chart
pname()
age_gender_group = data.groupby(['age', 'gender']).size().unstack()
age_gender_group.plot(kind='bar', stacked=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age and Gender Distribution')
plt.legend(title='Gender')
plt.show()
```

```python
# Heatmap
pname()
age_gender_votes = data.pivot_table(values='votes', index='age', columns='gender', aggfunc='sum')
plt.imshow(age_gender_votes, cmap='YlGnBu', interpolation='nearest', aspect='auto')
plt.colorbar(label='Votes')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Votes by Age and Gender')
plt.show()
```
