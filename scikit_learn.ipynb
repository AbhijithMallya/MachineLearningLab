## KNN

```python
import numpy as np
import pandas as pd
from sklearn import neighbors , metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('car.data')
print(data.head())
data
```

```python
X = data[['buying','maint','safety']].values
y = data[['class']]
```

```python
print(X.shape)
print(y.shape)
print(X)
print(y)
```

There is a drawback as some values in the features are vhigh , low , mid , high etc. These are string and not integer. To feed into an ML algorithm we need to convert them to numbers

## Converting the data

```python
#X convertion using LabelEncoder require X to be in .values . ie  numpy
Le = LabelEncoder()
for i in range(len(X[0])):
    X[:,i] = Le.fit_transform(X[:,i])
print(X)
```

```python
# y convertion --> Label Mapping 
label_mapping ={
    'unacc':0,
    'acc':1,
    'good':2,
    'vgood':3
}

y['class'] = y['class'].map(label_mapping)
```

```python
y.sample(20)
```

```python

y = np.array(y)
print(y)
```

### Create Model

```python
knn_model = neighbors.KNeighborsClassifier(n_neighbors=25,weights='uniform')

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
knn_model.fit(X_train,y_train)
```

```python
prediction  = knn_model.predict(X_test)
# Compute and print classification report
# print(classification_report(y_test, y_prediction))
# # Plot non-normalized confusion matrix
# plt.figure()
```

```python
accuracy = metrics.accuracy_score(y_test,prediction)
print("Predictions : ",prediction)
print("Accuracy : ",accuracy)
```
