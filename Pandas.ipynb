### Importing Pandas library

### Creating a Pandas DataFrame

```python
data = {'Name': ['John', 'Emily', 'Michael'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
```

### Viewing the first few rows of a DataFrame

```python
df.head()
```

### Accessing a column in a DataFrame

```python
df['City']
```

### Filtering Rows based  on a condition

```python
filtered_df = df[df['Age'] > 30]
filtered_df
```

```python
### Sorting a DataFrame by a Column
```

```python
sorted_df = df.sort_values('Age')
sorted_df
```

```python
### Performing statistical calculations in a Column
```

```python
average_age = df['Age'].mean()
print("Average age :"+str(average_age))
```

```python
### Adding a new Column to a DataFrame
```

```python
df['Gender'] = ['Male', 'Female', 'Male']
```

```python
### Grouping and agregating data in a DataFrame
```

```python
grouped_df = df.groupby('Gender').mean()
grouped_df
```
