```python
pname()
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('dirty_election_data.csv')
print("Reading the csv file !")
```

### Display the data

```python
pname()
#print(data)
data
```

### Remove Duplicates

```python
pname()
data = data.drop_duplicates()
print("Data after removing duplicates : \n")
print(data)
```

### Replace with mean

```python
pname()
#data.fillna(data.mean(), inplace=True)
numeric_columns = data.select_dtypes(include='number').columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
data
```

```python
pname()
data.to_csv('cleaned_election_data.csv', index=False)
print("Save the cleaned data to a new CSV %%file")
```

### Visualization

```python
pname()
plt.hist(data['votes'])
plt.xlabel('Votes')
plt.ylabel('Frequency')
plt.title('Distribution of Votes')
plt.show()
```
